```python
# Filter rows where both agent_dep and agent_job are non-empty
dep_filtered = dep.dropna(subset=['agent_dep', 'agent_job'])  # Removes rows with NULL values
dep_filtered = dep_filtered[dep_filtered['agent_dep'].str.strip() != ""]  # Removes rows with empty strings in agent_dep
dep_filtered = dep_filtered[dep_filtered['agent_job'].str.strip() != ""]  # Removes rows with empty strings in agent_job
dep_filtered = dep_filtered.rename(columns={'agent': 'agent_name'})
# Compare the length of the original and filtered dataset
print(len(dep))
print(len(dep_filtered))
```

```python
# Load table into a DataFrame using the engine
wechatnn = pd.read_sql_table("agent_wechatnn", con=engine)
wechatnn
```
