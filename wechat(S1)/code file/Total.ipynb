```python
import re
import numpy as np
import pandas as pd
import collections
```

```python
data = pd.read_csv("total.csv")
data.head()
```

```python
data_0=data[~(data.Read==0)]
```

```python
data_0
```

```python
c=collections.Counter(data_0.Author)
c=c.most_common()
author=[]
count=[]
for i in c:
    if i[1]>35:
        author.append(i[0])
        count.append(i[1])
        
        
c #copy to create count.csv
```

```python
name=u'曾小纾'
a_id=[]
article=[]
read=[]
max_layers=[]
for index, row in data_0.iterrows():
    if row["Author"]==name:
        title=row["Title"]
        title="".join(title.split(' '))
        if title not in article:
            article.append(title)
            a_id.append(row["No"])
            read.append(row["Read"])
            max_layers.append(row["Max_layers"])
        else:
            i=article.index(title)
            read[i]=read[i]+row["Read"]
            max_layers[i]=max(max_layers[i],row["Max_layers"])
            
article_data=zip(a_id,read,max_layers)
```
